**DISTRIBUTION OF CALORIES COUNT ACROSS THE MENU.**

```python
cal = df["Calories"].describe()
print(cal)
```

```python
sns.histplot(data = df, x= "Calories", color = "pink")
plt.figure(figsize = (12,5))
plt.show()
```

*Most menu items have a calorie count within the range (188 to 282)*

**Exploration of the nutritional content(fat,protein and carbohydrates) of different items along with the food items that contain
highest amount of these nutrients.**

<span style= "color: red">**Fat**</span>

```python
total_fat = df["Total Fat"].describe()
print(total_fat)
```

```python
plt.figure(figsize = (12,4))
sns.histplot(data = df, x = "Total Fat", kde = True, color = "orange")
plt.xlabel("Total Fat")
plt.ylabel("Frequency")
plt.show()
```

```python
max_fat =df.sort_values("Total Fat").tail(10)
plt.figure(figsize = (8,5))
sns.barplot( data = max_fat, x= "Total Fat", y = "Item")
plt.show()
```

<span style= "color: brown">*Chicken McNuggets(40 peice) contains highest amount of fat.*</span>

<span style= "color: red">**Carbohydrates**</span>

```python
total_carbs = df["Carbohydrates"].describe()
print(total_carbs)
```

```python
plt.figure(figsize = (10,5))
sns.histplot(data = df, x = "Carbohydrates", kde = True,color= 'green')
plt.xlabel("Carbs")
plt.ylabel("Frequency")
plt.show()
```

<span style= "color: red">**Protein**</span>

```python
protein = df["Protein"].describe()
print(protein)
```

```python
plt.figure(figsize = (10,5))
sns.histplot(data = df, x= "Protein", kde = True, color= "brown")
plt.xlabel("Protein")
plt.ylabel("Frequency")
plt.show()
```

```python
max_protein = df.sort_values("Protein").tail(10)
plt.figure(figsize = (10,4))
sns.barplot( data= max_protein, x = "Total Fat", y = "Item")
plt.show()
```

<span style= "color: brown">*Chicken McNuggets(40 piece) contains highets amount of protein.*</span>

```python
sns.barplot(data= df,x= "Vitamin A (% Daily Value)", y = "Category")
plt.figure(figsize= (8,5))
plt.show()
```

<span style= "color: brown">*Salads are rich in vitamin A whereas Bevarages do not have enough vitamin.*</span>

```python
plt.figure(figsize = (8,5))
sns.barplot(data = df, x = "Iron (% Daily Value)", y = "Category")
plt.show()
```

<span style= "color: brown">*Beef and Proks are rich in Iron and there's very less amount of Iron is Beverages*</span>

```python
plt.figure(figsize = (8,3))
sns.barplot(data = df, x = "Protein", y= "Category")
plt.show()
```

<span style= "color: brown">*Chicken and Fish are rich in Protein whereas beverages are considered a low protein diet.*</span>

**Identification of healthy and less healthy menu options based on nutrical content.**

```python
less_healthy_item = df.loc[df["Calories"].idxmax()]
print(f"Item with highest calories: Item: {less_healthy_item['Item']} Total Calories: {less_healthy_item['Calories']}")
```

```python
healthy_item = df.loc[df["Calories"].idxmin()]
print(f"Item with lowest calories: \n Item: {healthy_item['Item']}\n Total Calories: {healthy_item['Calories']}")
```

<span style= "color:green">*Chicken McNuggets (40 peice) is a less healthy option containing 1880 Calrories whereas Diet Coke is the healthy item with 0 calorie*</span>


<span style= "color: blue">**Nutritional analysis can provide significant benefits to both McDonald’s customers and the organization itself:**</span>

<span style= "color: brown">**Benefits for Customers:**</span>
Informed Choices: Customers can make healthier choices by understanding the nutritional content of menu items. This is especially important for those 
with specific dietary needs or health goals.

-Transparency: Providing detailed nutritional information builds trust and transparency, showing that McDonald’s cares about the well-being of its 
customers.

-Customization: With access to nutritional data, customers can better customize their orders to fit their dietary preferences, such as opting for 
lower-calorie or lower-sodium options.
    
<span style= "color: green">**Benefits for McDonald’s:**</span>
Enhanced Reputation: Offering healthier options and being transparent about nutritional content can improve McDonald’s reputation, attracting
health-conscious customers who might have previously avoided fast food.

-Menu Development: Nutritional analysis helps McDonald’s identify which menu items are popular and align with health trends, guiding the development of new, healthier menu options.

-Customer Satisfaction: By meeting the dietary needs and preferences of a diverse customer base, McDonald’s can enhance overall customer satisfaction
and loyalty.

Overall, nutritional analysis supports McDonald’s commitment to providing quality food while catering to the evolving preferences and health needs 
of its customers.

--------------------------------------------------------------------------------------------------------------------------------

<span style= "color: blue">**McDonald’s has made several strides in improving the nutritional profile of their menu over the years, but there are always more opportunities for enhancement. Here are some recommendations:**</span>

<span style= "color: brown">**Increase Plant-Based Options:**</span> Expanding the range of plant-based items, such as veggie burgers, salads, and plant-based protein bowls, can cater to health-conscious consumers and those following vegetarian or vegan diets.

<span style= "color: green">**Reduce Sodium and Sugar:**</span> Further reducing the sodium and sugar content in their menu items, especially in sauces, dressings, and beverages, 
can significantly improve the overall nutritional profile.

<span style= "color: purple">**Whole Grain Alternatives:**</span> Offering whole grain buns, wraps, and other bread products can provide more fiber and nutrients compared to refined grains.

<span style= "color: brown">**Transparent Nutritional Information:**</span> Making nutritional information more accessible and transparent can help customers make informed choices. 
This includes clear labeling of calories, fat, sugar, and sodium content on menus and packaging.

<span style= "color: green">**Healthier Kids’ Meals:**</span> Continuing to improve the nutritional quality of Happy Meals by including more fruits, vegetables, and whole grains, and offering water or milk as default beverage options.

<span style="color: purple">**Healthy Cooking Methods:**</span> Incorporating healthier cooking methods, such as grilling instead of frying, can reduce the fat content in many menu items.
Portion Control: Offering smaller portion sizes for high-calorie items can help customers manage their calorie intake more effectively.

<span style="color: brown">**Innovative Ingredients:**</span> Experimenting with innovative ingredients like quinoa, kale, and avocado can add nutritional value and appeal to
health-conscious consumers.

*By implementing these strategies, McDonald’s can continue to enhance the nutritional quality of their menu and meet the evolving preferences of their 
customers.*
